**Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Reading an excel file dataset**

```python
df=pd.read_excel('Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=range(20),skipfooter=2)
```

```python
df
```

**Basic Line plot**

```python
years=list(range(1980,2014))
```

```python
df.set_index('OdName', inplace=True)
```

```python
df.loc['Haiti', years].plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

**Number of immigrants from india and china from 1980 to 2013**

```python
df.loc[['India','China'],years].transpose().plot(kind='line')
plt.title("Immigrants from india to china")
plt.xlabel("years")
plt.ylabel("immigrants")
plt.legend().remove()
plt.show()
```
